```python
import pandas as pd

from plotnine import *
```

```python
majors = pd.read_csv('https://github.com/briandk/2020-virtual-program-in-data-science/raw/master/data/college-majors.csv')
```

```python
majors
```

Which majors have the highest and lowest share of women?

```python
majors.sort_values(by = 'ShareWomen')
```

What happens if we plot the total number of majors vs. the share of women?

```python
(
    ggplot(
        aes(
            x = 'ShareWomen',
            y = "Total"
        ),
        data = majors
    ) +
    geom_point()
)
```
