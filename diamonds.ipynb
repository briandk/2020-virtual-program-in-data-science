```python
# Make inline plots vector graphics instead of raster graphics
# from IPython.display import set_matplotlib_formats
# set_matplotlib_formats('pdf', 'svg')

import pandas as pd
import seaborn as sns

from plotnine import *
```

```python
diamonds = pd.read_csv("https://github.com/briandk/2020-virtual-program-in-data-science/raw/master/data/diamonds.csv")
```

```python
diamonds
```

```python
sns.scatterplot(
    x = 'carat',
    y = 'price',  
    data = diamonds
)
```

```python
sns.jointplot(
    x = 'carat',
    y = 'price',
    data = diamonds
)
```

```python
sns.pairplot(diamonds)
```

```python
sns.relplot(
    x = 'carat',
    y = 'price',
    col = 'clarity',
    data = diamonds
)
```

```python
(
  ggplot(
    aes(
      x = 'carat',
      y = 'price',
    ),
    data = diamonds 
  ) 
  + geom_point()
  + facet_wrap('~ clarity')
  + theme_538()
)
```
