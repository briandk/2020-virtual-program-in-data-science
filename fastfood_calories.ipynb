```python
import pandas as pd

from plotnine import *
```

```python
fast_food = pd.read_csv("https://github.com/briandk/2020-virtual-program-in-data-science/raw/master/data/fastfood_calories.txt")

fast_food = fast_food.drop('Unnamed: 0', axis = 1)
fast_food
```

```python
fast_food.groupby('restaurant').mean().sort_values('calories', ascending = False)[['calories']]
```

```python
fast_food.groupby('restaurant').mean().sort_values('sodium', ascending = False)[['sodium']]
```

```python
(
    ggplot() +
    geom_point(
        aes(
            x = 'protein',
            y = 'total_fat',
            color = 'restaurant',
        ),
        data = fast_food
    ) +
    scale_color_brewer(type='qual') +
    facet_wrap('~ restaurant')
)
```

```python
(
    ggplot() +
    geom_point(
        aes(
            x = 'protein',
            y = 'cal_fat',
            color = 'restaurant',
        ),
        data = fast_food
    ) +
    scale_color_brewer(type='qual') +
    facet_wrap('~ restaurant')
)
```

```python
(
    ggplot() +
    geom_text(
        aes(
            x = 'protein',
            y = 'total_fat',
            label = 'item'
        ),
        data = fast_food.query('restaurant == "Chick Fil-A"')
    )
)
```
